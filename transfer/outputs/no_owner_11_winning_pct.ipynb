```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import unidecode
from other_funcs import *
```

```python
def import_data_from_vastaav(year, n_gws):
    year_range = f'20{year-1}-{year}'
    gw_df_list = []
    for i in range(1, n_gws+1):
        # gw_url = f"https://raw.githubusercontent.com/vaastav/Fantasy-Premier-League/refs/heads/master/data/{year_range}/gws/gw{i}.csv"
        gw_url = f"https://raw.githubusercontent.com/doug-ellis/Fantasy-Premier-League/refs/heads/doug-ellis/gw7/data/{year_range}/gws/gw{i}.csv"
        
        gw_df = pd.read_csv(gw_url, index_col=0)
        gw_df['gw'] = i
        gw_df_list.append(gw_df)

    gw_df = pd.concat(gw_df_list)
    return gw_df.reset_index()

def clean_name(name):
    name = name.replace(" ", "_")
    name = name.replace("-", "_")
    name = unidecode.unidecode(name)
    return name.strip().lower()
```

```python
def get_no_owner_11(gw):
    df = pd.read_csv(f'predictions/predicted_gw{gw}_simple.csv')
    no_owner = df.query('owner.isna()', engine='python')
    _, no_owner_11 = get_best_11_noadj(no_owner.sort_values('predicted_points', ascending=False))
    return no_owner_11['full_name']
```

```python
gw_df = import_data_from_vastaav(26, 27)
gw_df['full_name'] = gw_df['name'].apply(clean_name)
gw_df = gw_df.set_index('full_name')
```

```python
for gw in range(7, 28):
    no_owner_11 = get_no_owner_11(gw)
    gw_df_gw = gw_df.query('gw==@gw')
    no_owner_11_score = gw_df_gw.loc[no_owner_11, 'total_points'].sum()
    print(f'GW{gw} - {no_owner_11_score}')
```
