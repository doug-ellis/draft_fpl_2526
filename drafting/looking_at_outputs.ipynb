```python
import pandas as pd
```

```python
df = pd.read_csv('output\\drafted_players_real.csv', index_col=0)
df.loc['Ryan_Christie', 'picked_by'] = 0
df.loc['Iliman_Ndiaye', 'picked_by'] = 6
df.loc['Ethan_Wheatley', 'picked_by'] = 0
df.loc['Nicolas_Jackson', 'picked_by'] = 4
```

```python
picked_by_name_map = {1: 'Marcus',
                      2: 'Rory',
                      3: 'Doug',
                      4: 'Will',
                      5: 'Dave',
                      6: 'Lucas'}

df['picked_by_name'] = df['picked_by'].map(picked_by_name_map)
df
```

```python
starting_teams = []
for user in range(1,7):
    starting_teams.append(df.query('picked_by == @user'))
```

```python
for team in starting_teams:
    print(f"Team {team['picked_by_name'].iloc[0]} picks:")
    print(team['y_pred'].sum())
```
